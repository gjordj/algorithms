```python
def exercise_1():
    i = 2
    j = 3
    k = 4
    x = i + j + k
    i = x - i - j
    j = x - j - k
    k = x - i - k
    print(i, j, k)
    print("Variable 'i' is {first}, Variable 'j' is {second}, and Variable 'k' is {third}.".format(first=i,
                                                                                                   second=j,
                                                                                                   third=k))
    print('\n')


def exercise_2():
    for i in range(1, 8):
        for j in range(i):
            print(i, end='')
        print()
    print('\n')


def exercise_3():
    for i in range(1, 10):
        if i % 2 == 1:
            for j in range(int((10 - i) / 2) + 1): 
                print('-', end='')
            for j in range(i):  # same as exercise_2
                print(i, end='')
            for j in range(int((10 - i) / 2) + 1):
                print('-', end='')
            print()
    print('\n')


def exercise_4(const_value=4):
    wall = "+===+===+"
    line = "|   |   |"

    for i in range(const_value - 1):
        print(wall)
        for j in range(const_value):
            print(line)
    print(wall)
    print('\n')
```

```python
if __name__ == '__main__':
    print('Exercise 1:')
    exercise_1()
    print('Exercise 2:')
    exercise_2()
    print('Exercise 3:')
    exercise_3()
    print('Exercise 4:')
    exercise_4()
```
