```python
import random


# 1.Write a print_factor function that prints the factors of a given integer
# passed as a parameter to the function. The factors are separated by the word “and”.
# Assume that the integer is no less than 1. You may want to use a fencepost loop. For
# example, when taking 15 as an input, the function prints
# 1 and 2 and 3 and 5 and 15

def print_factor(number=15):
    """

    Args:
        number: integer number.

    Returns: factors of the number.

    """
    print('Exercise 1. \n')

    for num in range(1, number + 1):
        if num != number:
            if number % num == 0:
                print(num, 'and', end=' ')
        else:
            print(num)


# 2.Write a function called gcd that accepts two integers as parameters and returns
# the greatest common divisor (GCD) of the two numbers. The GCD of two integers a and
# b is the largest integer that is a factor of both a and b. One efficient way to compute the
# GCD of two numbers is to use Euclid’s algorithm, which states the following:
# GCD(a, b) = GCD(b, a % b)
# GCD(a, 0) = Absolute value of a

def gcd(int_1=10, int_2=14):
    """

    Args:
        int_1: first number to calculate the GCD.
        int_2: second number to calculate the GCD.

    Returns: int, the GCD

    """
    print('\nExercise 2. \n')

    while int_2:
        temp_1 = int_1
        int_1 = int_2
        int_2 = temp_1 % int_2
    print('The GCD is: ', int_1)


# 3.Write a function digit_sum that returns the sum of digits of a given integer as
# the input to the function. For example, digit_sum(555) returns 15;
# digit_sum(156) returns 12. For negative input, the function just returns the
# absolute value of the sum of digit. For example, digit_sum(-123) returns 6;
# digit_sum(-99) returns 18. Do not use string manipulation in this question.
# Instead, deal with the number directly.

def digit_sum(digits):
    """

    Args:
        digits: int number to sum its digits.

    Returns: int, sum of the digits.

    """
    print('\nExercise 3. \n')
    sum = 0
    digits = abs(digits)
    while digits != 0:
        sum += (digits % 10)
        digits //= 10
    print('The sum of digits is: ', sum)


# 4.Write a function called three_heads that repeatedly flips a coin until the results
# of the coin toss are three heads in a row. You should use random.randint to make it
# equally likely that a head or a tail will appear. Each time the coin is flipped, display H for
# heads or T for tails. When three heads in a row are flipped, the function should print a
# congratulatory message. Here is a possible output of a call to the function:
# T T H T T T H T H T H H H
# Three heads in a row!

def three_heads():
    """

    Returns: the results of a coin flipped until it gets 3 heads in a row.

    """
    print('\nExercise 4. \n')

    flips = []
    range_start = 0
    range_end = 1
    while True:
        flip_coin = random.randint(range_start, range_end)
        if flip_coin == 0:
            print('H', end=' ')
        elif flip_coin == 1:
            print('T', end=' ')
        flips.append(flip_coin)
        if flips[-3:] == [0, 0, 0]:
            print('\n')
            print("Three heads in a row!")
            break


# 5.Given three integers, we want to check if one of them is an average of the other
# two. Write a function named ave_point with three integers as input parameters. The
# function returns True if the above condition holds, otherwise returns False.
# For example, ave_point(6, 1, 11) should return True because 6 is the average of 1
# and 11; ave_point(1, 11, 6) should also return True. Instead, ave_point(2, 3,
# 6) returns False because no integer is an average of the other two. Note that the
# function returns True if the three integers are the same.
def ave_point(num_1, num_2, num_3):
    """

    Args:
        num_1: int
        num_2: int
        num_3: int

    Returns: message of whether a number is the average of the others.

    """
    print('\nExercise 5. \n')

    average_1 = (num_1 + num_2) / 2
    average_2 = (num_2 + num_3) / 2
    average_3 = (num_1 + num_3) / 2
    if average_1 == num_3 or average_2 == num_1 or average_3 == num_2:
        print('True, one of the numbers is the average of the others.')
        return True
    else:
        print('False, none of the numbers is the average of the others.')
        return False


def main():
    print_factor(84)
    gcd(32, 60)
    digit_sum(-22)
    three_heads()
    ave_point(2, 3, 6)


if __name__ == '__main__':
    main()
```
