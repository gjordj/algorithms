```python
import random


# 1. Write a maximum function returning the larger integer between two integers that are passed as
# parameters. E.g., if 5 and 9 are passed to the max function, it returns 9. Use the input function in a program that
# prompts the user to enter two integers, and the program will display the value of the greater integer between the
# two. (Note: your program should not use math.max() function.)

def maximum():
    """

    Returns: the greater integer introduced by the user.

    """
    print("1. Max Number:\n")
    num_1 = int(input("Enter first number: "))
    num_2 = int(input("Enter second number: "))
    if num_1 > num_2:
        print(num_1)
    elif num_1 < num_2:
        print(num_2)
    else:
        print("Same number.")
    print('\n')


# Write a program for simple math addition quizzes. The program first displays
# two 3-digit randomly generated numbers to be added, e.g., (you may want to use
# random.randint,so do not forget import random)
#  111
# +222
# Then the program allows the user to enter their answer to the quiz. If the answer is
# correct, the program displays ‘Congrats! It is correct.’ Otherwise, the program displays
# ‘Incorrect. The correct answer is XXX.’ Note that XXX here is the correct answer to the
# quiz. E.g., XXX should be 333 in the example above, if the user does not input a correct
# answer.
def random_with_N_digits(n):
    """

    Args:
        n: int, number of digits.

    Returns: int, random number with N digits.

    """
    range_start = 10 ** (n - 1)
    range_end = (10 ** n) - 1
    return random.randint(range_start, range_end)


def math_addition_quizzes():
    """

    Returns: result of the quizz made.

    """
    print("2. Math Quizz\n")
    num_1 = random_with_N_digits(3)
    num_2 = random_with_N_digits(3)

    print(' ', num_1)
    print('+', num_2, '\n')
    result = int(input("Enter result: "))

    if result == (num_1 + num_2):
        print('Congrats! It is correct.')

    else:
        print(f'{result} should be {num_1 + num_2}')
    print('\n')


# 3. Write a function, word_count, that counts the number of words for a given string. For example,
# word_count(“This is a string.”) returns 4; word_count(“hi!”) returns 1; word_count(“ This is also a string with
# wide space!”) returns 8. Your program should not use string.split().
def evaluate_truth_table(was_space, is_space, count):
    """

    Args:
        was_space: int, variable to know if the loop was inside a space.
        is_space: int, variable to know if the current analyzed character is a space.
        count: int, counts number of words.

    Returns:
        count: int, that counts number of words.

    """
    # keep in a word, no action needed, unnecessary condition, but just for clarity
    if was_space == 0 and is_space == 0:
        pass
    # end of word, getting into a new space
    elif was_space == 0 and is_space == 1:
        was_space = 1
    # new word, count word, end of previous space
    elif was_space == 1 and is_space == 0:
        count += 1
        was_space = 0
    # keep in a space, no action needed, unnecessary condition, but just for clarity
    elif was_space == 1 and is_space == 1:
        pass
    return was_space, is_space, count


def word_count():
    """

    Returns: Number of words written by the user.

    """
    print("3. Counting Words.\n")
    phrase = input("Enter words: ")
    # phrase = '   afsdf asdfj   asdñflkjasdf asdlfjsdfoijwfoij      asdfasdf'
    count = 0
    was_space = 1  # to count first word
    space = ' '
    for character in range(len(phrase)):
        if phrase[character] == space:
            is_space = 1
        if phrase[character] != space:
            is_space = 0
        was_space, is_space, count = evaluate_truth_table(was_space, is_space, count)
    print("Number of words:", count)
    print('\n')


def main():
    # 1. Write a maximum function returning the larger integer between two integers that are passed as
    # parameters. E.g., if 5 and 9 are passed to the max function, it returns 9. Use the input function in a program
    # that
    # prompts the user to enter two integers, and the program will display the value of the greater integer between the
    # two. (Note: your program should not use math.max() function.)
    maximum()
    # Write a program for simple math addition quizzes. The program first displays
    # two 3-digit randomly generated numbers to be added, e.g., (you may want to use
    # random.randint,so do not forget import random)
    #  111
    # +222
    # Then the program allows the user to enter their answer to the quiz. If the answer is
    # correct, the program displays ‘Congrats! It is correct.’ Otherwise, the program displays
    # ‘Incorrect. The correct answer is XXX.’ Note that XXX here is the correct answer to the
    # quiz. E.g., XXX should be 333 in the example above, if the user does not input a correct
    # answer.
    math_addition_quizzes()

    # 3. Write a function, word_count, that counts the number of words for a given string. For example,
    # word_count(“This is a string.”) returns 4; word_count(“hi!”) returns 1; word_count(“ This is also a string with
    # wide space!”) returns 8. Your program should not use string.split().
    word_count()
```
