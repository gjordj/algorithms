```python
# 2.Write code that accepts a list of strings as its parameter and indicates whether
# that list is a palindrome. A palindrome list is one that reads the same forward as
# backward. For example, the list ["alpha", "beta", "gamma", "delta",
# "gamma", "beta", "alpha"] is a palindrome.

def palindrome_list(list_strings=['a', 'b', 'c', 'b', 'a']):
    print("Exercise 1:")
    if not list_strings:
        print("There is no list of strings.")
        print('\n')
        return
    check_palindrome = []
    for item in range(0, len(list_strings)):
        if list_strings[item] == list_strings[len(list_strings) - 1 - item]:
            check_palindrome.append(True)
        else:
            check_palindrome.append(False)
    if False in check_palindrome:
        print('False, the following list is not a palindrome list: ', list_strings)
        print('\n')
        return False
    else:
        print('True, the following list is a palindrome list:', list_strings)
        print('\n')
        return True


# 3.Write a function called scale_by_k that accepts a list of integers as a
# parameter and replaces every integer of value k with k copies of itself. For example, if
# the list stores the values [4, 1, 2, 0, 3] before the function is called, it should
# store the values [4, 4, 4, 4, 1, 2, 2, 3, 3, 3] after the function finishes
# executing. You should directly work on the given list instead of creating another list.
# Zeros and negative numbers should be removed from the list by this function.

def scale_by_k(num_list=[2, 1, 2, -1, 3, 0]):
    print("Exercise 2:")
    position = 0
    while position < len(num_list):
        int_to_copy = num_list[position]
        if int_to_copy > 0:
            for num_copies in range(1, int_to_copy):
                num_list.insert(position + num_copies, int_to_copy)
            position += int_to_copy
        else:
            del num_list[position]
    print(num_list, '\n')
    return num_list


# 4.Write a function called magic_square that accepts a two-dimensional list of
# integers (i.e., a matrix) as a parameter and returns True if all of the matrixâ€™s row sum,
# column sum, diagonal sum, and off-diagonal sum are equal. For example,
# [[2,7,6],[9,5,1],[4,3,8]] satisfies the condition and the function returns
# true as all 8 sums are 15 (row sum: 2+7+6, 9+5+1, 2+9+4; column sum: 7+5+3, 6+1+8,
# 6+1+8; diagonal sum: 2+5+8; off-diagonal sum: 6+5+4).

def create_dict_output(dictionary, type_sum, values):
    type_sum = type_sum + " Sums"
    if type_sum in dictionary:
        dictionary[type_sum].append(values)
    else:
        dictionary[type_sum] = [values]
    return dictionary


def magic_square(matrix=[[2, 7, 6], [9, 5, 1], [4, 3, 8]]):
    print("Exercise 3:")
    diagonal_1_values = []
    diagonal_2_values = []
    total_sum = []
    dictionary = {}
    for row in range(len(matrix)):
        # Rows Sum
        total_sum.append(sum(matrix[row]))
        dictionary = create_dict_output(dictionary, "Rows", sum(matrix[row]))
        # Diagonal 1 values.
        column = row
        diagonal_1_values.append(matrix[row][column])
        # Diagonal 2 values.
        column = len(matrix) - row - 1
        diagonal_2_values.append(matrix[row][column])
    # Diagonal Sums: To get the diagonal values we need to iterate through every row.
    total_sum.extend([sum(diagonal_1_values), sum(diagonal_2_values)])
    dictionary['Diagonal Sums'] = [sum(diagonal_1_values), sum(diagonal_2_values)]

    # Columns Sum
    for column in range(len(matrix)):
        columns_values = []
        # Columns values: to get the column values, we need to iterate through every row and column.
        for row in range(len(matrix)):
            columns_values.append(matrix[row][column])
        dictionary = create_dict_output(dictionary, "Columns", sum(columns_values))
        total_sum.append(sum(columns_values))

    print(dictionary)
    if len(set(total_sum)) == 1:
        print("True, all sums are the same.")
        return True
    else:
        print("False, all sums are not the same.")
        return False


def main():
    palindrome_list()
    scale_by_k()

    matrix = [[1, 7, 6], [9, 5, 1], [4, 3, 8]]
    magic_square(matrix)


if __name__ == '__main__':
    main()
```
